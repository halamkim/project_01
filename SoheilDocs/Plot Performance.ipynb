```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
sp_500 = yf.download("^GSPC", start="2010-01-01", end="2020-12-31")
vbtlx = yf.download("VBTLX", start="2010-01-01", end="2020-12-31")
```

```python
sp_500
```

```python
sp_500['strat_return'] = sp_500["Close"].pct_change().dropna()
```

```python
sp_500
```

```python
sp_500['cum_return'] = (1+ sp_500['strat_return']).cumprod()
```

```python
sp_500.index = pd.to_datetime(sp_500.index).date
```

```python
sp_500
```

```python
vbtlx['strat_return'] = vbtlx["Close"].pct_change().dropna()
```

```python
vbtlx
```

```python
vbtlx['cum_return'] = (1+ vbtlx['strat_return']).cumprod()
```

```python
vbtlx.index = pd.to_datetime(vbtlx.index).date
```

```python
vbtlx
```

```python
# Plot the cumulative return as curves
plt.plot(sp_500["cum_return"], label="S&P 500", color='red', linewidth=2)
plt.plot(vbtlx ["cum_return"], label="VBTLX", color='blue', linewidth=2)
plt.xlabel("Year")
plt.ylabel("Cumulative Return")
plt.title("Cumulative  Return of S&P 500 and VBTLX (2010-2020)")
plt.legend()
plt.grid(True)
plt.show()
```
